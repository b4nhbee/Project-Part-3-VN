The procedure for collecting the data was collecting the population of the black-throated tit and the silver throat tit in the Dongzhai National Nature Reserve in China during the birdâ€™s breeding seasons from the years 2008 to 2022, excluding 2020 due to the global pandemic. The nests were located by following adult birds for potential nesting sites where the adults and nestlings (adolescents) were tagged with rings and measured for their characteristics, including a blood sample to test for sex.
